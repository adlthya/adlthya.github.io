```python
import pandas as pd
import altair as alt
alt.data_transformers.enable("vegafusion")
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/building_inventory.csv")
```

```python
city_count = df.groupby('City').agg("sum").reset_index()
city_count.nlargest(5, 'Square Footage')
```

```python
df = df[(df['Year Constructed'] > 0) & 
        ((df['City'] == 'Chicago') | 
         (df['City'] == 'Springfield') |
         (df['City'] == 'Urbana') |
         (df['City'] == 'Carbondale') |
         (df['City'] == 'Champaign'))]
```

```python
df.City.unique()
```

```python
brush = alt.selection_interval()
```

```python
c1 = alt.Chart(df).mark_point().encode(
    x = alt.X("Year Constructed:Q", scale = alt.Scale(domain=[1750, 2020])),
    y = alt.Y("Square Footage:Q").scale(type='log'),
    color = alt.condition(brush, 'City:N', alt.value('lightgray')), 
).add_params(brush)

c2 = alt.Chart(df).mark_bar().encode(
    x = "City:N",
    y = "count()",
    color = "City:N"
).transform_filter(brush)
```
